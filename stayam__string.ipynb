```python
def uppercase_half_string(s: str) -> str:
    mid = len(s) // 2
    first_half = s[:mid]
    second_half = s[mid:].upper()
    return first_half + second_half
text = "programming"
print(uppercase_half_string(text))
```

que 8: Capitalize first and last character of each word

```python
def capitalize_first_last(s: str) -> str:
    words = s.split()
    result = []
    for word in words:
        if len(word) == 1:
            result.append(word.upper())
        else:
            new_word = word[0].upper() + word[1:-1] + word[-1].upper()
            result.append(new_word)
    return " ".join(result)
text = "python is awesome"
print(capitalize_first_last(text))
```

Que 14: Least Frequent Character

```python
from collections import Counter

def least_frequent_char(s: str) -> str:
    freq = Counter(s)
    min_count = min(freq.values())
    least_chars = [ch for ch, count in freq.items() if count == min_count]
    return least_chars
text = "successes"
print("Least frequent character(s):", least_frequent_char(text))
```

20: Generating random strings until a given string is generated


```python
import random
import string

def generate_until_match(target: str):
    letters = string.ascii_lowercase
    count = 0

    while True:
        count += 1
        rand_str = ''.join(random.choice(letters) for _ in range(len(target)))
        if rand_str == target:
            print(f"Match found after {count} attempts → {rand_str}")
            break
generate_until_match("abc")
```

26: Find uncommon words from two Strings

```python
def uncommon_words(str1, str2):
    words1 = set(str1.split())
    words2 = set(str2.split())
    return list(words1.symmetric_difference(words2))  # words unique to each string
s1 = "the quick brown fox"
s2 = "the quick white cat"
print(uncommon_words(s1, s2))
```

45: Generate possible valid IP addresses from given string

```python
def restore_ip_addresses(s):
    def is_valid(segment):
        return len(segment) == 1 or (segment[0] != '0' and int(segment) <= 255)

    res = []
    n = len(s)
    for i in range(1, min(4, n - 2)):
        for j in range(i + 1, min(i + 4, n - 1)):
            for k in range(j + 1, min(j + 4, n)):
                p1, p2, p3, p4 = s[:i], s[i:j], s[j:k], s[k:]
                if all(map(is_valid, [p1, p2, p3, p4])):
                    res.append(f"{p1}.{p2}.{p3}.{p4}")
    return res
s = "25525511135"
print(restore_ip_addresses(s))
```

3: Consecutive characters frequency


```python
def consecutive_frequency(s):
    if not s:
        return

    count = 1
    result = []

    for i in range(1, len(s)):
        if s[i] == s[i-1]:
            count += 1
        else:
            result.append((s[i-1], count))
            count = 1
    result.append((s[-1], count))
    return result
s = "aaabbcdddde"
freq = consecutive_frequency(s)
print(freq)
```

5: Check if a string can become empty by recursive deletion

```python
def can_become_empty(s, patterns):
    if s == "":
        return True

    for p in patterns:
        idx = s.find(p)
        while idx != -1:

            new_s = s[:idx] + s[idx+len(p):]
            if can_become_empty(new_s, patterns):
                return True
            idx = s.find(p, idx + 1)

    return False
s = "abccba"
patterns = ["ab", "cc", "ba"]
print(can_become_empty(s, patterns))

```

6: Minimum number of rotations to obtain a string

```python
def min_rotations(s1, s2):
    if len(s1) != len(s2):
        return -1

    double_s1 = s1 + s1
    if s2 not in double_s1:
        return -1
    index = double_s1.find(s2)
    return index
s1 = "abcde"
s2 = "cdeab"
print(min_rotations(s1, s2))

s1 = "abcde"
s2 = "abced"
print(min_rotations(s1, s2))
```

9: Sort String by K frequency

```python
from collections import Counter

def sort_by_k_frequency(s, k=None):
    freq = Counter(s)

    sorted_chars = sorted(freq.items(), key=lambda x: (-x[1], x[0]))

    if k:
        sorted_chars = sorted_chars[:k]
    result = ''.join(char * count for char, count in sorted_chars)
    return result
s = "tree"
print(sort_by_k_frequency(s))
print(sort_by_k_frequency(s, k=2))
```

12: Character position of Kth word

```python
def kth_word_position(s, k):
    words = s.split()
    if k < 1 or k > len(words):
        return -1

    position = 0
    for i in range(k-1):
        position += len(words[i]) + 1
    return position
s = "Python is fun to learn"
k = 3
print(kth_word_position(s, k))
```

13: Right and Left Shift characters

```python
def left_shift(s, n):
    n = n % len(s)
    return s[n:] + s[:n]

def right_shift(s, n):
    n = n % len(s)
    return s[-n:] + s[:-n]

s = "Python"
print(left_shift(s, 2))
print(right_shift(s, 2))
```

23: Remove K length Duplicates

```python
def remove_consecutive_k_duplicates(s, k):
    i = 0
    result = []
    while i < len(s):
        if i + k <= len(s) and len(set(s[i:i+k])) == 1:
            i += k
        else:
            result.append(s[i])
            i += 1
    return ''.join(result)
s = "aaabbccddd"
k = 2
print(remove_consecutive_k_duplicates(s, k))
```

26: Replace duplicate Occurrence

```python
def replace_duplicates(s, replacement='_'):
    seen = set()
    result = []
    for char in s:
        if char in seen:
            result.append(replacement)
        else:
            seen.add(char)
            result.append(char)
    return ''.join(result)
s = "programming"
print(replace_duplicates(s))
```
